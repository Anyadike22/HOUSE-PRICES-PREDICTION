```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#LOAD THE DATA
```

```python
data = pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
```

```python
#DROP NULL VALUES
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
# PRE-PROCESSING
```

```python
data
```

```python
#NORMALIZATION OF SKEWED FEATURES
```

```python
data['total_rooms'] = np.log(data['total_rooms'] + 1)
data['total_bedrooms'] = np.log(data['total_bedrooms'] + 1)
data['population'] = np.log(data['population'] + 1)
data['households'] = np.log(data['households'] + 1)
```

```python
data.hist(figsize = (15, 8))
```

```python
# ONE-HOT ENCODING OF THE CATEGORICAL DATA
```

```python
data
```

```python
new_data = data.join(pd.get_dummies(data.ocean_proximity).astype(int))
```

```python
new_data
```

```python
# CORRELATION OF THE NEW ENTIRE DATASET (AFTER ONE-HOT ENCODING)
```

```python
new_data = data.join(pd.get_dummies(data.ocean_proximity).astype(int)).drop(['ocean_proximity'], axis=1)
```

```python
new_data
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(new_data.corr(), annot=True, cmap="YlGnBu")
```

```python
plt.figure(figsize= (15, 8))
sns.scatterplot(x='latitude', y='longitude', data=new_data, hue='median_house_value', palette='coolwarm')
```
